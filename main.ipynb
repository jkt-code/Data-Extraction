```python
#Importing necessary libraries
import requests 
import csv
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.chrome.service import Service 
from webdriver_manager.chrome import ChromeDriverManager 
```

```python
#Link to extract
url = 'https://www.thriftbooks.com/'
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install())) 
driver.get(url) # Wait for the content to load 
driver.implicitly_wait(10) # Adjust the wait time if necessary 

# Extract data using Selenium best_selling_section
best_selling_section = driver.find_element(By.ID,'Recommendations')
books = best_selling_section.find_elements(By.CLASS_NAME,'BookSlideDesktop-Container')

#Extract Book details and storing them in Excel
with open('books.csv', 'w', newline='') as file: 
    writer = csv.writer(file) 
    writer.writerow(['Title','Author','Category','Price']) # Excel header 
    for n in range(5):
        titles = best_selling_section.find_elements(By.ID,f'recSection{n}')# Extracting book category
        for section in titles: 
            title = section.find_elements(By.CLASS_NAME,'BookSlide-Title')
            author = section.find_elements(By.CLASS_NAME,'BookSlide-Author')
            price = section.find_elements(By.CLASS_NAME,'BookSlide-Price')
            for t,a,p in zip(title,author,price):
                category = section.find_element(By.TAG_NAME, 'h2').text.strip()
                title_ = t.text.strip() 
                author_ = a.text.strip()
                price_ = p.text.strip()
                writer.writerow([title_,author_,category,price_]) 
                print(f"Book Title: {title_},Book Author: {author_},Book Category: {category},Book Price: {price_}")
        
driver.quit()
```
